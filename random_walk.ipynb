```python
#let us generate random numbers or a random walk

import random
import plot_style 
import matplotlib.pyplot as plt
import numpy as np
#import progress module for a progress bar
from tqdm import tqdm


def random_walk_1d(n_numbers):
    #random numbers between 0 and 1
    r_random = [random.random() for i in range(n_numbers)]
    change = [1 if i > 0.5 else -1 for i in r_random]
    #make cumulative sum
    change = [sum(change[:i]) for i in range(1, n_numbers)]
    return change


steps = 1000
n_walkers = 10000

def n_randoms_walks_1d(n_walks, n_numbers):
    walks = [random_walk_1d(n_numbers) for i in tqdm(range(n_walks), desc='Generating walks')]
    return walks

walks = n_randoms_walks_1d(n_walkers, steps) 
#n plots of the walks

#calulate number of crossinng with 0 
 

    
```

```python
walks.T
def n_return(walks):
    a  = walks.T
    crossing_time = np.zeros(len(a))
    for i in tqdm(range(len(a))):
        b = a[:i]
        zeros = np.sum(b == 0)
        crossing_time[i] = zeros
        
    return crossing_time

returns = n_return(walks)        
    
    
plt.plot(returns)  
    
```

```python
def power_law(x, a, b):
    return a*x**b

#make fit of the returns

from scipy.optimize import curve_fit

popt, pcov = curve_fit(power_law, time, returns)

plt.plot(time, power_law(time, *popt), label=f'Fit with a*x^b, , a, = b={popt[0]:.2f}')
plt.plot(time, returns, label='Returns to 0')
plt.legend()
plt.show()
```
