```python
# Import necessary libraries
import os
import pandas as pd
import re

# Define the path where the data is stored
path = 'data english football'

# List all files in the directory
files = os.listdir(path)

# Initialize lists to store the data
read_files = []
bet365_data = []
other_data = []

# Read each CSV file
for file in files[:-2]:
    # Read the CSV file
    df = pd.read_csv(os.path.join(path, file))
    
    # Identify columns that contain 'B365' within the first 4 characters
    bet365_columns = [col for col in df.columns if re.match(r'^.{0,3}B365', col)]
    
    # Split the data into Bet365-related columns and other columns
    if bet365_columns:
        # Store Bet365 data in one dataframe
        bet365_data.append(df[bet365_columns])
    
    # Store all non-Bet365 columns in another dataframe
    other_columns = [col for col in df.columns if col not in bet365_columns]
    other_data.append(df[other_columns])

# Combine all Bet365 data and all other data into single DataFrames
bet365_combined = pd.concat(bet365_data, ignore_index=True)
other_combined = pd.concat(other_data, ignore_index=True)


other_combined = other_combined.fillna(np.inf)

bet365_combined = bet365_combined.fillna(np.inf)


```

```python
#in this section we make our y data and turn letters into binary values
results = np.zeros((len(bet365_combined),14))


for i in range(len(bet365_combined)):
    if other_combined['FTR'][i] == 'H':
        results[i][0] = 1
        results[i][1] = 0
        results[i][2] = 0
    if other_combined['FTR'][i] == 'D':
        results[i][0] = 0
        results[i][1] = 1
        results[i][2] = 0
    if other_combined['FTR'][i] == 'A':
        results[i][0] = 0
        results[i][1] = 0
        results[i][2] = 1
    #find total number of goals
    
    if other_combined['FTHG'][i] + other_combined['FTAG'][i] > 2.5:
        results[i][3] = 1
        results[i][4] = 0
    else:
        results[i][3] = 0
        results[i][4] = 1
    #calculate if the Asian handicap is won
    
    if other_combined['FTHG'][i] - other_combined['FTAG'][i] > bet365_combined['B365AHH'][i]:
        results[i][5] = 1
        results[i][6] = 0
    else:
        results[i][5] = 0
        results[i][6] = 1
    #calculate if the corner bet is won
    if other_combined['HC'][i] > other_combined['AC'][i]:
        results[i][7] = 1
        results[i][8] = 0
        results[i][9] = 0
    if other_combined['HC'][i] < other_combined['AC'][i]:
        results[i][7] = 0
        results[i][8] = 1
        results[i][9] = 0
    if other_combined['HC'][i] == other_combined['AC'][i]:
        results[i][7] = 0
        results[i][8] = 0
        results[i][9] = 1
    #calculate if the over 2.5 goals in corner betting is won
    if other_combined['HC'][i] + other_combined['AC'][i] > 2.5:
        results[i][10] = 1
        results[i][11] = 0
    else:
        results[i][10] = 0
        results[i][11] = 1
    #asian handicap for corner betting
    if other_combined['HC'][i] - other_combined['AC'][i] > bet365_combined['B365CAHH'][i]:
        results[i][12] = 1
        results[i][13] = 0
    else:
        results[i][12] = 0
        results[i][13] = 1

    
#calculate the potential profit for each bet

profit = np.zeros((len(bet365_combined),14))

mask = results > 0.5

for i in range(len(bet365_combined)):
    for j in range(14):
        if mask[i][j] == True:
            profit[i][j] = bet365_combined.iloc[i][j]

```

```python
def plot_function(arr1, arr2, title, x_label, y_label):
    plt.plot(arr1, arr2)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
def plot_scatter(arr1, arr2, title, x_label, y_label):
    plt.scatter(arr1, arr2)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
def plot_hist(arr, title, x_label, y_label):
    x_min = np.min(arr)
    x_max = np.max(arr)
    n_bins = int(np.sqrt(len(arr)))
    bins = np.linspace(x_min, x_max, n_bins)
    plt.hist(arr, bins=bins)
    plt.title(title)
    plt.xlabel(x_label)
    plt.ylabel(y_label)
    plt.show()
```
